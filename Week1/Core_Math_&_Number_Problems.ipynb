```python
import math
def sum_of_digits(n):
    n = abs(n)
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total

n = int(input("Enter the number: "))
print(sum_of_digits(n))
```

```python
def product_of_digits(n):
    n = abs(n)
    product = 1

    if n == 0:
        return 0

    while n > 0:
        product *= n % 10
        n //= 10

    return product


n = int(input("Enter the number: "))
print(product_of_digits(n))
```

```python
def reverse_number(n):
    sign = -1 if n < 0 else 1
    n = abs(n)
    rev = 0

    while n > 0:
        rev = rev * 10 + (n % 10)
        n //= 10

    return sign * rev


n = int(input("Enter the number: "))
print(reverse_number(n))
```

```python
def is_palindrome(n):
    original = n
    n = abs(n)
    rev = 0

    while n > 0:
        rev = rev * 10 + (n % 10)
        n //= 10

    return original == rev


n = int(input("Enter the number: "))
if is_palindrome(n):
    print("Palindrome number")
else:
    print("Not a palindrome number")
```

```python
def is_armstrong(n):
    if n < 0:
        return False

    temp = n
    digits = 0
    while temp > 0:
        digits += 1
        temp //= 10

    temp = n
    total = 0
    while temp > 0:
        digit = temp % 10
        total += digit ** digits
        temp //= 10

    return total == n


n = int(input("Enter the number: "))
if is_armstrong(n):
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```

```python
def is_perfect(n):
    if n <= 1:
        return False

    total = 0
    for i in range(1, n // 2 + 1):
        if n % i == 0:
            total += i

    return total == n


n = int(input("Enter the number: "))
if is_perfect(n):
    print("Perfect number")
else:
    print("Not a perfect number")
```

```python
def is_strong(n):
    if n < 0:
        return False

    temp = n
    total = 0

    while temp > 0:
        digit = temp % 10

        fact = 1
        for i in range(1, digit + 1):
            fact *= i

        total += fact
        temp //= 10

    return total == n


n = int(input("Enter the number: "))
if is_strong(n):
    print("Strong number")
else:
    print("Not a strong number")
```

```python
def is_harshad(n):
    if n <= 0:
        return False

    temp = n
    digit_sum = 0

    while temp > 0:
        digit_sum += temp % 10
        temp //= 10

    return n % digit_sum == 0


n = int(input("Enter the number: "))
if is_harshad(n):
    print("Harshad (Niven) number")
else:
    print("Not a Harshad number")
```

```python
def is_automorphic(n):
    if n < 0:
        return False

    square = n * n
    temp = n

    while temp > 0:
        if square % 10 != temp % 10:
            return False
        square //= 10
        temp //= 10

    return True


n = int(input("Enter the number: "))
if is_automorphic(n):
    print("Automorphic number")
else:
    print("Not an Automorphic number")
```

```python
def is_neon(n):
    if n < 0:
        return False

    square = n * n
    digit_sum = 0

    while square > 0:
        digit_sum += square % 10
        square //= 10

    return digit_sum == n


n = int(input("Enter the number: "))
if is_neon(n):
    print("Neon number")
else:
    print("Not a Neon number")
```

```python
def gcd(a, b):
    a = abs(a)
    b = abs(b)

    while b != 0:
        a, b = b, a % b

    return a


def lcm(a, b):
    if a == 0 or b == 0:
        return 0
    return abs(a * b) // gcd(a, b)


a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("LCM =", lcm(a, b))
```

```python
def power(a, b):
    result = 1

    if b < 0:
        a = 1 / a
        b = -b

    for _ in range(b):
        result *= a

    return result


a = int(input("Enter base (a): "))
b = int(input("Enter exponent (b): "))
print("Result =", power(a, b))
```

```python
def power_mod(a, b, m):
    result = 1
    a = a % m

    while b > 0:
        if b % 2 == 1:        # if b is odd
            result = (result * a) % m
        a = (a * a) % m
        b //= 2

    return result


a = int(input("Enter a: "))
b = int(input("Enter b: "))
m = int(input("Enter m: "))

print("Result =", power_mod(a, b, m))
```

```python
def prime_factors(n):
    n = abs(n)
    factors = []

    while n % 2 == 0:
        factors.append(2)
        n //= 2

    i = 3
    while i * i <= n:
        while n % i == 0:
            factors.append(i)
            n //= i
        i += 2


    if n > 1:
        factors.append(n)

    return factors


n = int(input("Enter the number: "))
print("Prime factors:", prime_factors(n))
```

```python
def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False

    i = 3
    while i * i <= n:
        if n % i == 0:
            return False
        i += 2

    return True


n = int(input("Enter the number: "))
if is_prime(n):
    print("Prime number")
else:
    print("Not a prime number")
```

```python
def sieve(n):
    if n < 2:
        return []

    is_prime = [True] * (n + 1)
    is_prime[0] = is_prime[1] = False

    p = 2
    while p * p <= n:
        if is_prime[p]:
            for multiple in range(p * p, n + 1, p):
                is_prime[multiple] = False
        p += 1

    primes = []
    for i in range(2, n + 1):
        if is_prime[i]:
            primes.append(i)

    return primes


n = int(input("Enter upper limit: "))
print("Primes:", sieve(n))
```

```python
def twin_primes(n):
    if n < 3:
        return []

    is_prime = [True] * (n + 1)
    is_prime[0] = is_prime[1] = False

    p = 2
    while p * p <= n:
        if is_prime[p]:
            for multiple in range(p * p, n + 1, p):
                is_prime[multiple] = False
        p += 1

    twins = []
    for i in range(3, n - 1):
        if is_prime[i] and is_prime[i + 2]:
            twins.append((i, i + 2))

    return twins


n = int(input("Enter upper limit: "))
print("Twin primes:", twin_primes(n))
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)


def nCr(n, r):
    if r < 0 or r > n:
        return 0

    r = min(r, n - r)
    return factorial(n) // (factorial(r) * factorial(n - r))


n = int(input("Enter n: "))
r = int(input("Enter r: "))

print("nCr =", nCr(n, r))
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)


def nPr(n, r):
    if r < 0 or r > n:
        return 0
    return factorial(n) // factorial(n - r)


n = int(input("Enter n: "))
r = int(input("Enter r: "))

print("nPr =", nPr(n, r))
```

```python
def harmonic_sum(n):
    if n <= 0:
        return 0

    total = 0.0
    for i in range(1, n + 1):
        total += 1 / i

    return total


n = int(input("Enter n: "))
print("Sum =", harmonic_sum(n))
```

```python
def sin_taylor(x, n):
    term = x
    sinx = x
    sign = -1
    power = x * x * x
    fact = 6

    for i in range(1, n):
        sinx += sign * (power / fact)
        sign *= -1
        power *= x * x
        fact *= (2*i + 2) * (2*i + 3)

    return sinx


x = float(input("Enter x (in radians): "))
n = int(input("Enter number of terms: "))

print("sin(x) =", sin_taylor(x, n))
```

```python
def exp_taylor(x, n):
    term = 1.0   # first term (x^0 / 0!)
    result = 1.0

    for i in range(1, n):
        term *= x / i
        result += term

    return result


x = float(input("Enter x: "))
n = int(input("Enter number of terms: "))

print("e^x =", exp_taylor(x, n))
```
